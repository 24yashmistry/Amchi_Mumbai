# Outlier Detection and Removal

```python
data.describe()
```

```python
data['Address'].value_counts()
```

```python
data = data[data['Bathrooms'] >= 1]
data.describe()
```

```python
def remove_outlier_psqft(df):
    df_output = pd.DataFrame()
    for key, subdf in df.groupby('Address'):
        m = np.mean(subdf.Price_sqft)
        st = np.std(subdf.Price_sqft)
        
        gen_df = subdf[(subdf.Price_sqft > (m-st)) & (subdf.Price_sqft < (m+st))]
        df_output = pd.concat([df_output,gen_df], ignore_index = True)
    
    return df_output

data = remove_outlier_psqft(data)
```

```python
data.describe()
```

```python
data.shape
```

```python
def bedroom_outlier_remover(df):
    exclude_indices = np.array([])
    for add, add_df in df.groupby('Address'):
        bed_stats = {}
        for bed, bed_df in add_df.groupby('Bedrooms'):
            bed_stats[bed] = {
                'mean' : np.mean(bed_df.Price_sqft),
                'std' : np.std(bed_df.Price_sqft),
                'count' : bed_df.shape[0]
            }
#         print(add, bed_stats)
        
        for bed, bed_df in add_df.groupby('Bedrooms'):
            stats = bed_stats.get(bed-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bed_df[bed_df.Price_sqft < (stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')
        
data = bedroom_outlier_remover(data)
```

```python
data.shape
```

```python
def bathroom_outlier_remover(df):
    exclude_indices = np.array([])
    for add, add_df in df.groupby('Address'):
        bath_stats = {}
        for bath, bath_df in add_df.groupby('Bathrooms'):
            bath_stats[bath] = {
                'mean' : np.mean(bath_df.Price_sqft),
                'std' : np.std(bath_df.Price_sqft),
                'count' : bath_df.shape[0]
            }
        
        for bath, bath_df in add_df.groupby('Bathrooms'):
            stats = bath_stats.get(bath-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bath_df[bath_df.Price_sqft < (stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')
        
data = bathroom_outlier_remover(data)
```

# Model training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
from sklearn.neural_network import MLPRegressor

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
X = data.drop(columns=['price'])
y = data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False), ['Address']), remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
# y_pred_lr
```

```python
r2_score(y_test, y_pred_lr)
```

```python
lasso = Lasso()
pipe = make_pipeline(column_trans, scaler, lasso)
pipe.fit(X_train, y_train)
y_pred_lasso = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_lasso)
```

```python
fr = RandomForestRegressor()
pipe = make_pipeline(column_trans, scaler, fr)
pipe.fit(X_train, y_train)
y_pred_fr = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_fr)
```

```python
mlp = MLPRegressor()
pipe = make_pipeline(column_trans, scaler, mlp)
pipe.fit(X_train, y_train)
y_pred_mlp = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_mlp)
```

```python
rd = Ridge()
pipe = make_pipeline(column_trans, scaler, rd)
pipe.fit(X_train, y_train)
y_pred_rd = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_rd)
```

```python
gbr = GradientBoostingRegressor()
pipe = make_pipeline(column_trans, scaler, gbr)
pipe.fit(X_train, y_train)
y_pred_gbr = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_gbr)
```

# Gradient Boosting Regressor had best score

```python
print(r2_score(y_test, y_pred_lr))
print(r2_score(y_test, y_pred_lasso))
print(r2_score(y_test, y_pred_rd))
print(r2_score(y_test, y_pred_fr))
print(r2_score(y_test, y_pred_gbr))
print(r2_score(y_test, y_pred_mlp))
```
