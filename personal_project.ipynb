```python
def commaremoval(x):
    x = x.replace(',','')
    return x
```

```python
data['Address'] = data['Address'].apply(commaremoval)
```

```python
areas = ['Andheri', 'Bandra', 'Borivali', 'Dahisar', 'Goregaon', 'Jogeshwari', 'Juhu', 'Kandivali', 'Khar', 'Malad', 'Santacruz', 'Vile', 'Bhandup', 'Chandivali', 'Govandi', 'Ghatkopar', 'Kanjurmarg', 'Kanjur', 'Kurla', 'Mulund', 'Nahur', 'Powai', 'Vidyavihar', 'Vikhroli', 'Chembur', 'Mankhurd', 'Wadala', 'Trombay', 'Antop', 'Bhuleshwar', 'Byculla', 'Colaba', 'Cumbala', 'Dadar', 'Girgaon', 'Kalbadevi', 'Kamathipura', 'Malabar', 'Matunga', 'Parel', 'Sion', 'Tardeo', 'Umarkhadi', 'Dharavi', 'Thakur', 'Yari', 'Worli', 'Mira', 'Eksar', 'South', 'Sewri', 'Marol', 'Banganga']
def convert(x):
    temp = x.split()
    for i in temp:
        i = i.capitalize()
        if i in areas:
            return i
```

```python
data['Address'] = data['Address'].apply(convert)
```
